```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
import datetime as dt
import seaborn as sns

df = pd.read_csv('employment.csv')
df.head()
```

```python
# Question 1 

df['datestamp'] = pd.to_datetime(df['datestamp'])
df.set_index(df.datestamp)
print('Number of missing values values in each column: ')
df.isna().sum(axis='rows')
```

```python
# Question 2

df.plot(figsize=(30,20), kind='box')
```

```python
# Question 3
df_temp = df.drop(['datestamp'], axis = 1)
df.head()
df.plot(figsize=(30,20), x = 'datestamp', y = df_temp.columns, kind='line')
```

```python
# Question 4
print('Monthly trends for the unemployment: ')
print(df.plot(figsize=(10,10), subplots = True, x = 'datestamp',kind='line'))
```

```python
print('Yearly trends for the unemployment: ')
df['year'] = df.datestamp.dt.year
new_df = pd.DataFrame(df.groupby(['year']).mean())
new_df.plot(figsize=(10,10),kind='line')
```

```python
# Question 5
import statsmodels.api as sm
from pylab import rcParams

rcParams['figure.figsize'] = 18, 8
decomposition = sm.tsa.seasonal_decompose(new_df, model='additive')
fig = decomposition.plot()
plt.show()
```
