```python
import pandas as pd
from numpy import percentile
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('mtcars.csv')
df
```

```python
# finding the best mpg 

bestmpg = df.iloc[df['mpg'].idxmax(), 0:]
print('Best mpg:', bestmpg[0], ',mpg:', bestmpg[1])

# finding the worst mpg

worstmpg = df.iloc[df['mpg'].idxmin(), 0:]
print('Worst mpg:', worstmpg[0], ',mpg:', worstmpg[1])

# best horsepower

besthp = df.iloc[df['hp'].idxmax(), 0:]
print('Most Horsepower:', besthp[0], ',Horspower:', besthp[4])
```

```python
# 5 number summary of displacement 

quartiles = percentile(df['disp'], [25, 50, 75])
minDisp = df.iloc[df['disp'].idxmin(), 0:]
maxDisp = df.iloc[df['disp'].idxmax(), 0:]

print('Minimum Value of displacement: ', minDisp[3])
print('Maximum Value of displacement: ', maxDisp[3])
print('First Quartile:', quartiles[0])
print('Median:', quartiles[1])
print('Third Queartile:', quartiles[2])

df['disp'].describe()
```

```python
# Find the median horse power
print('Median of horsepower:', df['hp'].median())
```

```python
# Finding the manual vs automatic mpg

print('mpg of manual cars:', df.loc[df['am'] == 0, 'mpg'].mean())
print('mpg of automatic cars:', df.loc[df['am'] == 1, 'mpg'].mean())
```

```python
# drawing histogram of miles per gallon 

plt.hist(df['mpg'], bins = 20)
plt.show()
```

```python
# box plot for mpg and number of cylinders 

cyl4 = df.loc[df['cyl'] == 4, 'mpg']
cyl6 = df.loc[df['cyl'] == 6, 'mpg']
cyl8 = df.loc[df['cyl'] == 8, 'mpg']

cyl_data = [cyl4, cyl6, cyl8]
plt.boxplot(cyl_data)

plt.show()
```

```python
# ploting crosstable for number of automatic vs manual cars 

manualCount = df.loc[df['am'] == 0, 'mpg'].count()
autoCount = df.loc[df['am'] == 1, 'mpg'].count()
print(manualCount, autoCount)
pd.crosstab(index = df['am'],columns = ['manual', 'automatic'])
```

```python
# plotting crosstable for automatic to cylinders 

pd.crosstab(index = df['am'], columns = df['cyl'])
```

```python
# correlation between weight and mpg

print(df['wt'].corr(df['mpg']))
```
